## xml

```python
import urllib3
import requests
import xml.etree.ElementTree as et
```

```python
LINK = "http://www.chilkatsoft.com/xml-samples/nutrition.xml"
response = requests.get(LINK)
with open('bookstore.xml', 'wb') as file:
    file.write(response.content)
filename = 'bookstore.xml'
file = open(filename, mode='r')
text = file.read()
print(file.closed)
file.close()
print(text)
```
